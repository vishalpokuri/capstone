```python
import pandas as pd
import matplotlib.pyplot as plt

baseline1 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.9107, 0.9483, 1.0000]
}

baseline2 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.8625, 0.9345, 0.9513]
}

baseline3 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.8376, 0.9410, 0.9006]
}

baseline4 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.9150, 0.9551, 0.9299]
}

baseline5 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.8598, 0.9520, 0.9337]
}

baselines = {
    "ResNet18 + ProtoNet": pd.DataFrame(baseline1),
    "ResNet50 + ProtoNet": pd.DataFrame(baseline2),
    "VGG16 + ProtoNet": pd.DataFrame(baseline3),
    "MS-ProtoNet-18": pd.DataFrame(baseline4),
    "MS-ProtoNet-50": pd.DataFrame(baseline5),
}

plt.figure()

for name, df in baselines.items():
    plt.plot(df["Shots"], df["Accuracy"], marker='o', label=name)

plt.xlabel("Number of Shots")
plt.ylabel("Episodic test accuracy")
plt.title("Standard Few-Shot Episodic Test Accuracy")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

baseline1 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.7756, 0.7867, 0.9132]
}

baseline2 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.7500, 0.7645, 0.7865]
}

baseline3 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.7253, 0.6959, 0.7563]
}

baseline4 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.7981, 0.7969, 0.7665]
}

baseline5 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.7736, 0.8545, 0.7986]
}

baselines = {
   "ResNet18 + ProtoNet": (baseline1),
    "ResNet50 + ProtoNet": (baseline2),
    "VGG16 + ProtoNet": (baseline3),
    "MS-ProtoNet-18": (baseline4),
    "MS-ProtoNet-50": (baseline5),
}

shots = [1, 5, 10]
x = np.arange(len(shots))
bar_width = 0.15

plt.figure()

for i, (name, data) in enumerate(baselines.items()):
    plt.bar(
        x + i * bar_width,
        data["Accuracy"],
        width=bar_width,
        label=name
    )

plt.xlabel("Number of Shots")
plt.ylabel("Full-test robustness accuracy")
plt.title("Full-Test Robustness Accuracy Across Baselines")
plt.xticks(x + bar_width * 2, shots)
plt.legend()
plt.show()
```
