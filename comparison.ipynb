```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

baseline1 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.8690, 0.9745, 0.9700]
}

baseline2 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.8595, 0.9625, 0.9650]
}

baseline3 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.8710, 0.9595, 0.9740]
}

baseline4 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.9150, 0.9860, 0.9350]
}

baseline5 = {
    'Shots': [1, 5, 10],
    'Accuracy': [0.9015, 0.9445, 0.9730]
}

baselines = {
   "ResNet18 + ProtoNet": (baseline1),
    "ResNet50 + ProtoNet": (baseline2),
    "VGG16 + ProtoNet": (baseline3),
    "MS-ProtoNet-18": (baseline4),
    "MS-ProtoNet-50": (baseline5),
}

shots = [1, 5, 10]
x = np.arange(len(shots))
bar_width = 0.15

plt.figure()

for i, (name, data) in enumerate(baselines.items()):
    plt.bar(
        x + i * bar_width,
        data["Accuracy"],
        width=bar_width,
        label=name
    )

plt.xlabel("Number of Shots")
plt.ylabel("Accuracy")
plt.title("Accuracy Comparison Across Baselines (Bar Chart)")
plt.xticks(x + bar_width * 2, shots)
plt.legend()
plt.show()
```
